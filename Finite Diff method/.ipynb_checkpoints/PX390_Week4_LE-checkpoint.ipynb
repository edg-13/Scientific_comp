```python
import matplotlib as mpl
#mpl.use("nbagg")
import matplotlib.pyplot as plt
import numpy as np
```

# Example 1

## No damping
# $x(t) = x_0 + v_{x,0} t$
# $y(t) = y_0 + v_{y,0} t - \frac{1}{2}g t^2$

## with damping
# $ x(t) = x_0 + \frac{1-\exp(-\frac{\gamma}{m} t)}{\gamma}m v_{x,0} $
# $ y(t) = y_0 + \frac{1-\exp(-\frac{\gamma}{m} t)}{\gamma}m v_{y,0} -\frac{m}{\gamma}g m t + \frac{1-\exp(-\frac{\gamma}{m} t)}{\gamma^2}m^2 g$

```python
# x0 = 0.0
# y0 = 0.0
# vx0 = 10.0
# vy0 = 0.0
# gamma = 2.0
# dt = 0.01


x0 = 10.0
y0 = 10.0
vx0 = 1.25
vy0 = 5.75
gamma = 2.0
dt = 1.0

with open("/Users/albert/Google Drive/PX390_Week4_LE/le4_1_input.txt","w") as f:
    f.write("{} {} {} {} {} {}".format(x0,y0,vx0,vy0,gamma,dt))
```

```python
out = np.loadtxt("/Users/albert/Google Drive/PX390_Week4_LE/output.dat")

t = out[:,0]
m = 1.0
g = 9.80665

if gamma > 0.0:
    my_x = x0 + (1 - np.exp(-gamma*t/m))*m*vx0/gamma
    my_y = -( np.exp(-gamma*t/m) - 1 )*g*m**2/gamma**2 + y0 - g*m*t/gamma + (1 - np.exp(-gamma*t/m))*m*vy0/gamma 
else:
    my_x = x0+vx0*t
    my_y = y0+vy0*t-0.5*g*t**2
```

```python
plt.plot(t,out[:,1])
plt.plot(t,my_x,"--")
plt.xlabel("time (s)")
plt.ylabel("x (m)")
plt.show()

plt.plot(t,out[:,1]-my_x)
plt.xlabel("time (s)")
plt.ylabel("error in x (m)")
plt.show()

plt.semilogy(t,np.abs(out[:,1]-my_x))
plt.xlabel("time (s)")
plt.ylabel("absolute error in x (m)")
plt.show()

plt.plot(t,out[:,2])
plt.plot(t,my_y,"--")
plt.xlabel("time (s)")
plt.ylabel("y (m)")
plt.show()

plt.plot(t,out[:,2]-my_y)
plt.xlabel("time (s)")
plt.ylabel("error in y (m)")
plt.show()

plt.semilogy(t,np.abs(out[:,2]-my_y))
plt.xlabel("time (s)")
plt.ylabel("absolute error in y (m)")
plt.show()
```

# Example 2

# $\frac{\mathrm{d}y}{\mathrm{d}x} = \gamma y$
# $y(x) = y_0 \exp( \gamma x )$

```python
out = np.loadtxt("/Users/albert/Google Drive/PX390_Week4_LE/output_2.dat")
```

```python
t = out[:,0]

y0 = 1.0
gamma = -4.0
my_y = y0*np.exp(gamma*t)
```

```python
plt.plot(t,out[:,1])
plt.plot(t,my_y,"--")
plt.xlabel("time (s)")
plt.ylabel("y (a.u.)")
plt.show()

plt.plot(t,out[:,1]-my_y)
plt.xlabel("time (s)")
plt.ylabel("error in y (a.u.)")
plt.show()

plt.semilogy(t,np.abs(out[:,1]-my_y))
plt.xlabel("time (s)")
plt.ylabel("absolute error in y (a.u.)")
plt.show()
```
