```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive
```

```python
L = 10
N = 100
D = 0.1
v = 1.0
k_plus = 1.0
k_minus = 0.5
with open("C:/Users/44748/Documents/Uni/Year 3 Stuff/Scientific Computing/Assignment 4/input.txt","w") as f:
    f.write("{} {} {} {} {} {}".format(L, N, D, v, k_plus, k_minus))
```

```python
out = np.loadtxt("C:/Users/44748/Documents/Uni/Year 3 Stuff/Scientific Computing/Assignment 4/results.txt")
```

```python
def f(t=0):
    fig, axs = plt.subplots(2, sharex=True, gridspec_kw={"hspace":0.5},figsize=(5,5))
    axs[0].plot(out[:N, 0], out[(100*N*t):(100*N*t)+N, 2])
    #axs[0].plot(out[:N, 0], out[t:t+N,2])
    axs[0].set_title("Concentration of reactant A")
    axs[1].plot(out[:N, 0], out[(100*N*t):(100*N*t)+N, 3])
    #axs[1].plot(out[:N, 0], out[t:t+N,3])
    axs[1].set_title("Concentration of reactant B")
    plt.show()

interactive_plot = interactive(f, t=(0, 59))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
#f(1)
```
